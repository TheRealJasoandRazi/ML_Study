```python
!pip install kagglehub
import kagglehub

# Download latest version
path = kagglehub.dataset_download("srgiomanhes/steam-games-dataset-2025")

print("Path to dataset files:", path)
```

```python
import pandas as pd

full_path = path + "/steam_games.csv"
data = pd.read_csv(full_path)

# Display the first few rows of the DataFrame
print(data.head())
```

```python
# now that we have the data loaded lets explore it
import matplotlib.pyplot as plt

numerical_features = data.select_dtypes(include=['number']).columns

# Create a plot for each feature against 'price'
for feature in numerical_features:
    plt.figure(figsize=(5, 4))  # Set the size of each plot
    plt.scatter(data[feature], data['price_initial (USD)'], color='blue', alpha=0.5)  # Scatter plot
    plt.title(f'{feature} vs Price')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.grid(True)
    
    plt.xlim(0, data[feature].max())
    plt.ylim(0, 100)
            
    plt.show()
```
