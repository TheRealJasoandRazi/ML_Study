```python
# Now lets get a simple random forest model trained on all of the data (before feature extraction/selection, scaling, etc)
# create a pipeline that does one hot encoding for the categorical data
```

```python
numerical_features = [x for x in numerical_features if x != 'Price']
```

```python
# split price into its own panda dataframe
X = data.drop(columns=["Price"])  # Assuming "price" is the target variable
y = data["Price"]
X
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# pipeline
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),  # Fill missing values with median
    ("scaler", StandardScaler())  # Standardize numerical features
])

# Categorical Transformer: Handle missing values & encode
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),  # Fill missing values with most frequent value
    ("encoder", OneHotEncoder(handle_unknown="ignore"))  # One-hot encode categorical features
])

# Combine transformations
preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, numerical_features),
    ("cat", categorical_transformer, categorical_features)
])
```

```python
#prepocess data
X_preprocessed = preprocessor.fit_transform(X)
```

```python
# split data into three sets (training, testing, validation)
X_train, X_test_and_val, y_train, y_test_and_val = train_test_split(X_preprocessed, y, test_size=0.3, random_state=42)
X_test, X_val, y_test, y_val = train_test_split(X_test_and_val, y_test_and_val, test_size=0.5, random_state=42)
```

```python
type(X_train)
```

```python
# now that the data is ready, lets train out first random forest
from sklearn.ensemble import RandomForestRegressor

default_regressor = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
default_regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

y_pred = default_regressor.predict(X_test)

# Evaluate the model using MSE
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"Mean Dollar Error: {rmse}")
```

```python
# aleks metric for a good model, is less than 10% off

# Calculate percentage error for each prediction
def calculate_error_range(y_pred, y_test): 
    percentage_error = np.abs((y_test - y_pred) / y_test) * 100
    print(percentage_error[:10])  # Show the first 10 values
    mean_percentage_error = np.mean(percentage_error)
    print(f'Mean Percentage Error: {mean_percentage_error:.2f}%')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
import numpy as np

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'n_estimators': [50, 100, 200],  # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],  # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],  # Minimum number of samples required to be at a leaf node
    'max_features': ['sqrt', 'log2'],  # Number of features to consider when looking for the best split
}

rf = RandomForestRegressor(random_state=42, n_jobs=-1)

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error')
grid_search.fit(X_train_selected_features, y_train)

best_params = grid_search.best_params_
best_rf = grid_search.best_estimator_

print(f"Best hyperparameters: {best_params}")
```

```python
# lets test the best model on the test set!
y_pred = best_rf.predict(X_test_selected_features)
calculate_error_range(y_pred, y_test)
```

```python
# the decision tree seems to perform pretty bad so ill leave it out
# run grid search on the linear regression and the knn regressor to get optimal hyper parameters

from sklearn.linear_model import Ridge

param_grid_lr = {
    'alpha': [0.1, 1, 10, 100, 1000]  # Regularization strength
}

# Create Ridge model (for regularization)
ridge = Ridge()

# Perform GridSearchCV
grid_search_lr = GridSearchCV(estimator=ridge, param_grid=param_grid_lr, cv=5)
grid_search_lr.fit(X_train_selected_features, y_train)

# Get the best model and parameters
best_lr = grid_search_lr.best_estimator_
best_lr_params = grid_search_lr.best_params_

print(f"Best hyperparameters for Linear Regression: {best_lr_params}")
```

```python
# got best linear regressor now find the best knn regressor
knn_regressor = KNeighborsRegressor()

# Define the parameter grid
param_grid = {
    'n_neighbors': [3, 5, 7, 10, 15],          # Number of neighbors to test
    'weights': ['uniform', 'distance'],       # Weighting function for neighbors
    'p': [1, 2]                              # Distance metric (1 = Manhattan, 2 = Euclidean)
}

# Perform GridSearchCV
grid_search_knn = GridSearchCV(knn_regressor, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search_knn.fit(X_train_selected_features, y_train)

# Get the best parameters and best score
best_knn_params = grid_search_knn.best_params_
best_knn_score = grid_search_knn.best_score_
```

```python
best_knn_regressor = KNeighborsRegressor(n_neighbors=best_knn_params['n_neighbors'],
                                         weights=best_knn_params['weights'],
                                         p=best_knn_params['p'])
best_knn_regressor.fit(X_train_selected_features, y_train)
y_pred = best_knn_regressor.predict(X_test_selected_features)
calculate_error_range(y_pred, y_test)
```
