```python
df = pd.read_csv(path + "/water_potability.csv")
df
```

```python
# clean the data
# check for missing data, if there is missing data then fill with median value

df.isnull().sum()
```

```python
median_value = df.median()
df_filled_median = df.fillna(median_value)
df_filled_median
```

```python
# drop duplicates
unique_df = df_filled_median.drop_duplicates()
unique_df
```

```python
# use boxplots to find outliers
import matplotlib.pyplot as plt

unique_columns = unique_df.columns.tolist()

fig, ax = plt.subplots()
for i in range(len(unique_columns)):
    ax = unique_df.boxplot(column=unique_columns[i])
    plt.title(f'Box Plot of Column {unique_columns[i]}')
    plt.show()
```

```python
# replace outliers with the boundry value, keeping samples should help the model train better
import numpy as np
from scipy import stats

clean_df = unique_df.copy()
for col in unique_columns:
    # find upper and lower quartile for each column
    Q1 = clean_df[col].quantile(0.25) 
    Q3 = clean_df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    upperBoundry = Q3 + 1.5 * IQR
    lowerBoundry = Q1 - 1.5 * IQR
    
    # if any values pass those boundaries for the given column, then set them to the boundry
    clean_df.loc[clean_df[col] < lowerBoundry, col] = lowerBoundry
    clean_df.loc[clean_df[col] > upperBoundry, col] = upperBoundry
```

```python
# check if outliers have been replaced by running box plots again
fig, ax = plt.subplots()
for i in range(len(unique_columns) - 1):
    ax = clean_df.boxplot(column=unique_columns[i])
    plt.title(f'Box Plot of Column {unique_columns[i]}')
    plt.show()
```

```python
# Studio 2
# count the distributions of potability (target variable) and ensure that there is a equal amount of 0s and 1s

clean_df['Potability'].value_counts()
```

```python
# there is a significant difference of 700 samples between the two classes, this means that the dataset is skewed
# and could possibly lead to a biased model as a result
```

```python
# normalise the dataset (except the target variable) using min max scaling
def normalise_dataset(dataset):
    for column in dataset.columns: 
        if column != "Potability": # avoid normalising the target variable
            dataset[column] = (dataset[column] - dataset[column].min()) / (dataset[column].max() - dataset[column].min()) 
    return dataset

df_min_max_scaled = normalise_dataset(clean_df.copy())
print(df_min_max_scaled)
```

```python
# Since no features have any noticible relationships feature engineering will be difficult
# the strongest relationship is between solids and sulfate at -0.14

#create raw dataset
y = clean_df["Potability"] 
raw_clean_df = clean_df.drop(columns=["Potability"])
raw_clean_df = raw_clean_df.astype('float32')

# create new features (explain why these in report)
feature_engineered_df = raw_clean_df.copy()
feature_engineered_df['Hardness_ph_Ratio'] = feature_engineered_df['Hardness'] / feature_engineered_df['ph']
feature_engineered_df['Solids_ph_Ratio'] = feature_engineered_df['Solids'] / (feature_engineered_df['ph'])
feature_engineered_df['Sulfate_Hardness_Product'] = feature_engineered_df['Sulfate'] * feature_engineered_df['Hardness']
feature_engineered_df['Chloramines_Solids_Difference'] = feature_engineered_df['Chloramines'] - feature_engineered_df['Solids']
feature_engineered_df['Sulfate_Solids_Ratio'] = feature_engineered_df['Sulfate'] / (feature_engineered_df['Solids'])
```

```python
# create 5 different sets of features to train model on
# ask teacher about feature engineering when all features have no relationships
from sklearn.preprocessing import PolynomialFeatures

list_of_datasets = []

raw_df = raw_clean_df.copy()
list_of_datasets.append(["raw_df",raw_df])

raw_df_normalised = normalise_dataset(raw_df.copy())
list_of_datasets.append(["raw_df_normalised",raw_df_normalised])

list_of_datasets.append(["feature_engineered_df",feature_engineered_df])

# some of the features i created had infinity values
feature_engineered_df.replace([np.inf, -np.inf], np.nan, inplace=True)
feature_engineered_df.fillna(feature_engineered_df.mean(), inplace=True)

feature_engineered_df_normalised = normalise_dataset(feature_engineered_df.copy())
list_of_datasets.append(["feature_engineered_df_normalised",feature_engineered_df_normalised])

poly = PolynomialFeatures(degree=2)
poly_features = poly.fit_transform(raw_df.copy())
poly_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(raw_df.copy().columns))
list_of_datasets.append(["poly_df",poly_df])

poly_df_normalised = normalise_dataset(poly_df.copy())
list_of_datasets.append(["poly_df_normalised",poly_df_normalised])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

def train_and_test_model_on_dataset(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) 

    clf = DecisionTreeClassifier(random_state=42)
    clf.fit(X_train, y_train)

    y_pred = clf.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy:.4f}")

    return clf
```

```python
# create X and y for each dataset and train/test the model on each dataset
for i in range(len(list_of_datasets)):
    _tuple = list_of_datasets[i]
    df_name = _tuple[0]
    print(f"using dataset {df_name}")
    X = _tuple[1] 
    train_and_test_model_on_dataset(X, y)
```
