```python
import os
import pandas as pd

current_dir = os.getcwd()
data_path = current_dir + "/Data"
csv = os.listdir(data_path)
Boning_file_path = os.path.join(data_path, "Boning.csv")
Boning_df = pd.read_csv(Boning_file_path, index_col=None, header=0)
Slicing_file_path = os.path.join(data_path, "Slicing.csv")
Slicing_df = pd.read_csv(Slicing_file_path, index_col=None, header=0)
Boning_df.columns
```

```python
items_to_filter = [
     'Left Toe x', 
     'Left Toe y', 
     'Left Toe z', 
     'Right Toe x', 
     'Right Toe y', 
     'Right Toe z', 
     'Frame'
]
Boning_filtered_df = Boning_df.filter(items_to_filter)
Slicing_filtered_df = Slicing_df.filter(items_to_filter)
```

```python
Slicing_filtered_df
```

```python
# add target variable and combine two dataframes
Slicing_filtered_df['Target'] = 1
Boning_filtered_df['Target'] = 0
Combined_df = pd.concat([Slicing_filtered_df, Boning_filtered_df], axis=0)
Combined_df
```

```python
# get root mean square value of combined features
import numpy as np

# Write function for the root mean square
def append_root_mean_square_feature(dataset, column1, column2):
    column1_np = dataset[column1].to_numpy()
    column2_np = dataset[column2].to_numpy()
    root_mean_square = [(np.sqrt(np.mean(x**2)) + np.sqrt(np.mean(y**2))) / 2 for x, y in zip(column1_np, column2_np)]
    new_column_name = f"RMS of {column1} and {column2}"
    dataset[new_column_name] = root_mean_square
    return dataset

# same function as before but with three columns
def append_root_mean_square_feature_3(dataset, column1, column2, column3):
    column1_np = dataset[column1].to_numpy()
    column2_np = dataset[column2].to_numpy()
    column3_np = dataset[column3].to_numpy()
    root_mean_square = [
        (np.sqrt(np.mean(x**2)) + np.sqrt(np.mean(y**2)) + np.sqrt(np.mean(z**2))) / 3
        for x, y, z in zip(column1_np, column2_np, column3_np)
    ]
    new_column_name = f"RMS of {column1} and {column2} and {column3}"
    dataset[new_column_name] = root_mean_square
    return dataset

def append_roll(dataset, col1, col2, col3):
    col1_np = dataset[col1].to_numpy()
    col2_np = dataset[col2].to_numpy()
    col3_np = dataset[col3].to_numpy()
    roll = [
        (180 / np.pi) * np.arctan2(y, np.sqrt(x**2 + z**2))
        for x, y, z in zip(col1_np, col2_np, col3_np)
    ]
    new_column_name = f"Roll of {col1} and {col2} and {col3}"
    dataset[new_column_name] = roll
    return dataset

def append_pitch(dataset, col1, col2, col3):
    col1_np = dataset[col1].to_numpy()
    col2_np = dataset[col2].to_numpy()
    col3_np = dataset[col3].to_numpy()
    pitch = [
        (180 / np.pi) * np.arctan2(x, np.sqrt(y**2 + z**2))
        for x, y, z in zip(col1_np, col2_np, col3_np)
    ]
    new_column_name = f"Pitch of {col1} and {col2} and {col3}"
    dataset[new_column_name] = pitch
    return dataset
    
```

```python
Combined_df_expanded = Combined_df.copy()

# ROOT MEAN SQUARE CALCULATIONS
Combined_df_expanded = append_root_mean_square_feature(Combined_df_expanded, "Left Toe x", "Left Toe y")
Combined_df_expanded = append_root_mean_square_feature(Combined_df_expanded, "Right Toe x", "Right Toe y")

Combined_df_expanded = append_root_mean_square_feature(Combined_df_expanded, "Left Toe y", "Left Toe z")
Combined_df_expanded = append_root_mean_square_feature(Combined_df_expanded, "Right Toe y", "Right Toe z")

Combined_df_expanded = append_root_mean_square_feature(Combined_df_expanded, "Left Toe z", "Left Toe x")
Combined_df_expanded = append_root_mean_square_feature(Combined_df_expanded, "Right Toe z", "Right Toe x")

Combined_df_expanded = append_root_mean_square_feature_3(Combined_df_expanded, "Left Toe x", "Left Toe y", "Left Toe z")
Combined_df_expanded = append_root_mean_square_feature_3(Combined_df_expanded, "Right Toe x", "Right Toe y", "Right Toe z")

# ROLL AND PITCH CALCULATIONS

Combined_df_expanded = append_roll(Combined_df_expanded, "Left Toe x", "Left Toe y", "Left Toe z")
Combined_df_expanded = append_roll(Combined_df_expanded, "Right Toe x", "Right Toe y", "Right Toe z")

Combined_df_expanded = append_pitch(Combined_df_expanded, "Left Toe x", "Left Toe y", "Left Toe z")
Combined_df_expanded = append_pitch(Combined_df_expanded, "Right Toe x", "Right Toe y", "Right Toe z")

Combined_df_expanded
```

```python
# Step 3: Data pre-processing and Feature computation

def calculate_new_feature(dataset, no_of_frames, method, columns):
    method_to_call = f"{method}_operation"
    new_df = pd.DataFrame()
    for i in range(len(columns)): # loop through each column
        column = columns[i]
        if column == 'Target' or column == 'Frame': #only processes the 18 features
            continue
        col_np = dataset[column].to_numpy()
        result = globals()[method_to_call](col_np, no_of_frames)
        new_df[f'{method} of {column}'] = result
        
        # this block of code creates a new dataframe in one contiguous block, prevents memory fragmentation
        new_df_copy = new_df.copy()
        del new_df
        new_df = new_df_copy.copy()
        del new_df_copy
        
    print(len(new_df))
    return new_df

def mean_operation(col_np, no_of_frames):
    return np.mean(col_np.reshape(-1, no_of_frames), axis=1) # makes matrix with each row having 60 elements and returns the mean of each row

def standard_deviation_operation(col_np, no_of_frames):
    return np.std(col_np.reshape(-1, no_of_frames), axis=1)

def min_operation(col_np, no_of_frames):
    return np.min(col_np.reshape(-1, no_of_frames), axis=1)

def max_operation(col_np, no_of_frames):
    return np.max(col_np.reshape(-1, no_of_frames), axis=1)

def auc_operation(col_np, no_of_frames):
    blocks = col_np.reshape(-1, no_of_frames)
    auc = []
    for i in range(len(blocks)):
        block = blocks[i]
        calc = np.cumsum(block)-np.concatenate(([0], np.cumsum(np.diff(block)/2)), axis=0)
        # get mean of AUC
        auc.append(np.mean(calc))
    return auc

def peaks_operation(col_np, no_of_frames):
    blocks = col_np.reshape(-1, no_of_frames)
    res_arr = []
    
    for i in range(len(blocks)):
        block = blocks[i]
        res = 0
        for idx in range(1, len(block) - 1):
            if block[idx + 1] > block[idx] < block[idx - 1] or block[idx + 1] < block[idx] > [idx - 1]:
                res += 1
        res_arr.append(res)
        
    return res_arr
```

```python
original_columns = Combined_df_expanded.columns
print(f"original columns is {original_columns}")

list_of_features = []
list_of_features.append(calculate_new_feature(Combined_df_expanded, 60, "mean", original_columns))
list_of_features.append(calculate_new_feature(Combined_df_expanded, 60, "standard_deviation", original_columns))
list_of_features.append(calculate_new_feature(Combined_df_expanded, 60, "min", original_columns))
list_of_features.append(calculate_new_feature(Combined_df_expanded, 60, "max", original_columns))
list_of_features.append(calculate_new_feature(Combined_df_expanded, 60, "auc", original_columns))
list_of_features.append(calculate_new_feature(Combined_df_expanded, 60, "peaks", original_columns))
data = pd.concat(list_of_features, axis=1)
statistical_df = pd.DataFrame(data)
statistical_df
```

```python
blocks = np.arange(len(Combined_df['Target'])) // 60
target_values = Combined_df.groupby(blocks)['Target'].apply(lambda x: x.mode()[0]).to_numpy()
y = target_values
```

```python
# Step 4: Training

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.decomposition import PCA

X_train, X_test, y_train, y_test = train_test_split(statistical_df, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC

svm_1 = SVC(kernel='rbf', C=1, gamma='scale')
svm_1.fit(X_train, y_train)

svm_1.score(X_test, y_test)
```

```python
cross_val_score(svm_1, X_train, y_train, cv=10, n_jobs=-1) 
```

```python
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 'auto', 0.001, 0.0001],
    'kernel': ['rbf']
}

grid_search = GridSearchCV(SVC(), param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)

print(grid_search.best_params_)
print(grid_search.best_score_)
```

```python
svm_2 = grid_search.best_estimator_
svm_2.fit(X_train, y_train)
svm_2.score(X_test, y_test)
```

```python
cross_val_score(svm_2, X_train, y_train, cv=10, n_jobs=-1) 
```

```python
selector = SelectKBest(score_func=f_classif, k=10)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

grid_search.fit(X_train_selected, y_train)

print(X_train.shape)
print(X_train_selected.shape)
```

```python
svm_3 = grid_search.best_estimator_
svm_3.fit(X_train_selected, y_train)
svm_3.score(X_test_selected, y_test)
```

```python
cross_val_score(svm_3, X_train_selected, y_train, cv=10, n_jobs=-1) 
```

```python
pca = PCA(n_components=10)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)

grid_search.fit(X_train_pca, y_train)

print(X_train.shape)
print(X_train_pca.shape)
```

```python
svm_4 = grid_search.best_estimator_
svm_4.fit(X_train_pca, y_train)
svm_4.score(X_test_pca, y_test)
```

```python
cross_val_score(svm_4, X_train_pca, y_train, cv=10, n_jobs=-1) 
```
