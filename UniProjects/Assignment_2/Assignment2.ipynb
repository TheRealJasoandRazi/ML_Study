```python
import os
import pandas as pd

current_dir = os.getcwd()
data_path = current_dir + "/Data"
csv = os.listdir(data_path)
Boning_file_path = os.path.join(data_path, "Boning.csv")
Boning_df = pd.read_csv(Boning_file_path, index_col=None, header=0)
Slicing_file_path = os.path.join(data_path, "Slicing.csv")
Slicing_df = pd.read_csv(Slicing_file_path, index_col=None, header=0)
Boning_df.columns
```

```python
items_to_filter = [
     'Left Toe x', 
     'Left Toe y', 
     'Left Toe z', 
     'Right Toe x', 
     'Right Toe y', 
     'Right Toe z', 
     'Frame'
]
Boning_filtered_df = Boning_df.filter(items_to_filter)
Slicing_filtered_df = Slicing_df.filter(items_to_filter)
```

```python
Slicing_filtered_df
```

```python
# add target variable and combine two dataframes
Slicing_filtered_df['Target'] = 1
Boning_filtered_df['Target'] = 0
Combined_df = pd.concat([Slicing_filtered_df, Boning_filtered_df], axis=0)
Combined_df
```

```python
# get root mean square value of combined features
import numpy as np

# Write function for the root mean square
def append_root_mean_square_feature(dataset, column1, column2):
    column1_np = dataset[column1].to_numpy()
    column2_np = dataset[column2].to_numpy()
    root_mean_square = [(np.sqrt(np.mean(x**2)) + np.sqrt(np.mean(y**2))) / 2 for x, y in zip(column1_np, column2_np)]
    new_column_name = f"RMS of {column1} and {column2}"
    dataset[new_column_name] = root_mean_square
    return dataset

# same function as before but with three columns
def append_root_mean_square_feature_3(dataset, column1, column2, column3):
    column1_np = dataset[column1].to_numpy()
    column2_np = dataset[column2].to_numpy()
    column3_np = dataset[column3].to_numpy()
    root_mean_square = [
        (np.sqrt(np.mean(x**2)) + np.sqrt(np.mean(y**2)) + np.sqrt(np.mean(z**2))) / 3
        for x, y, z in zip(column1_np, column2_np, column3_np)
    ]
    new_column_name = f"RMS of {column1} and {column2} and {column3}"
    dataset[new_column_name] = root_mean_square
    return dataset

def append_roll(dataset, col1, col2, col3):
    col1_np = dataset[col1].to_numpy()
    col2_np = dataset[col2].to_numpy()
    col3_np = dataset[col3].to_numpy()
    roll = [
        (180 / np.pi) * np.arctan2(y, np.sqrt(x**2 + z**2))
        for x, y, z in zip(col1_np, col2_np, col3_np)
    ]
    new_column_name = f"Roll of {col1} and {col2} and {col3}"
    dataset[new_column_name] = roll
    return dataset

def append_pitch(dataset, col1, col2, col3):
    col1_np = dataset[col1].to_numpy()
    col2_np = dataset[col2].to_numpy()
    col3_np = dataset[col3].to_numpy()
    pitch = [
        (180 / np.pi) * np.arctan2(x, np.sqrt(y**2 + z**2))
        for x, y, z in zip(col1_np, col2_np, col3_np)
    ]
    new_column_name = f"Pitch of {col1} and {col2} and {col3}"
    dataset[new_column_name] = pitch
    return dataset
    
```

```python
Combined_df_expanded = Combined_df.copy()

# ROOT MEAN SQUARE CALCULATIONS
Combined_df_expanded = append_root_mean_square_feature(Combined_df_expanded, "Left Toe x", "Left Toe y")
Combined_df_expanded = append_root_mean_square_feature(Combined_df_expanded, "Right Toe x", "Right Toe y")

Combined_df_expanded = append_root_mean_square_feature(Combined_df_expanded, "Left Toe y", "Left Toe z")
Combined_df_expanded = append_root_mean_square_feature(Combined_df_expanded, "Right Toe y", "Right Toe z")

Combined_df_expanded = append_root_mean_square_feature(Combined_df_expanded, "Left Toe z", "Left Toe x")
Combined_df_expanded = append_root_mean_square_feature(Combined_df_expanded, "Right Toe z", "Right Toe x")

Combined_df_expanded = append_root_mean_square_feature_3(Combined_df_expanded, "Left Toe x", "Left Toe y", "Left Toe z")
Combined_df_expanded = append_root_mean_square_feature_3(Combined_df_expanded, "Right Toe x", "Right Toe y", "Right Toe z")

# ROLL AND PITCH CALCULATIONS

Combined_df_expanded = append_roll(Combined_df_expanded, "Left Toe x", "Left Toe y", "Left Toe z")
Combined_df_expanded = append_roll(Combined_df_expanded, "Right Toe x", "Right Toe y", "Right Toe z")

Combined_df_expanded = append_pitch(Combined_df_expanded, "Left Toe x", "Left Toe y", "Left Toe z")
Combined_df_expanded = append_pitch(Combined_df_expanded, "Right Toe x", "Right Toe y", "Right Toe z")

Combined_df_expanded
```

```python
# Step 3: Data pre-processing and Feature computation

Combined_df_expanded_2 = Combined_df_expanded.copy()

def calculate_new_feature(dataset, no_of_frames, method):
    method_to_call = f"{method}_operation"
    columns = dataset.columns
    for i in range(len(columns)): # loop through each column
        column = columns[i]
        if column == 'Target' or column == 'Frame': #only processes the 18 features
            continue
        col_np = dataset[column].to_numpy()
        result = globals()[method_to_call](col_np, no_of_frames)
        dataset[f'{method} of {column}'] = result
        
        # this block of code creates a new dataframe in one contiguous block, prevents memory fragmentation
        dataset_copy = dataset.copy()
        del dataset
        dataset = dataset_copy.copy()
        del dataset_copy
        
    return dataset

def mean_operation(col_np, no_of_frames):
    means = np.mean(col_np.reshape(-1, no_of_frames), axis=1) # makes matrix with each row having 60 elements and returns the mean of each row
    return np.repeat(means, no_of_frames)

def standard_deviation_operation(col_np, no_of_frames):
    std_devs = np.std(col_np.reshape(-1, no_of_frames), axis=1)
    return np.repeat(std_devs, no_of_frames)
```
