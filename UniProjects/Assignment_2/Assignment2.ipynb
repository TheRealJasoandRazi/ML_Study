```python
import os
import pandas as pd

current_dir = os.getcwd()
data_path = current_dir + "/Data"
csv = os.listdir(data_path)
Boning_file_path = os.path.join(data_path, "Boning.csv")
Boning_df = pd.read_csv(Boning_file_path, index_col=None, header=0)
Slicing_file_path = os.path.join(data_path, "Slicing.csv")
Slicing_df = pd.read_csv(Slicing_file_path, index_col=None, header=0)
Boning_df.columns
```

```python
items_to_filter = [
     'Left Toe x', 
     'Left Toe y', 
     'Left Toe z', 
     'Right Toe x', 
     'Right Toe y', 
     'Right Toe z', 
     'Frame'
]
Boning_filtered_df = Boning_df.filter(items_to_filter)
Slicing_filtered_df = Slicing_df.filter(items_to_filter)
```

```python
Slicing_filtered_df
```

```python
# add target variable and combine two dataframes
Slicing_filtered_df['Target'] = 1
Boning_filtered_df['Target'] = 0
Combined_df = pd.concat([Slicing_filtered_df, Boning_filtered_df], axis=0)
Combined_df
```

```python
# get root mean square value of combined features
import numpy as np

left_toe_x = Combined_df["Left Toe x"].to_numpy()
left_toe_y = Combined_df["Left Toe y"].to_numpy()
toe_x_y = [(np.sqrt(np.mean(x**2)) + np.sqrt(np.mean(y**2))) / 2 for x, y in zip(left_toe_x, left_toe_y)]
toe_x_y
```
