```python
def checkDataStructure(dir_path):
    total_size = 0
    file_count = 0
    for i in range(len(dir_path)):
        directory_file_count = 0
        for dirpath, dirnames, filenames in os.walk(dir_path[i]):
            for file in filenames:
                file_path = os.path.join(dirpath, file)
                try:
                    total_size += os.path.getsize(file_path)
                    file_count = file_count + 1
                    directory_file_count = directory_file_count + 1
                except PermissionError:
                    print(f"Skipping file due to permission error: {file_path}")

            print(directory_file_count)

    print(f"total_size is {total_size}")
    print(f"file_count is {file_count}")
```

```python
# the oob score is 0.6569 which is slightly worse than without the blur
# before we try PCA again i want to see if data augmentation can help

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
import numpy as np
from PIL import Image
import cv2

def resize_image(image_path, target_size=(224, 224)):
    image = Image.open(image_path).resize(target_size)
    return np.array(image)  # Convert to NumPy array inside this function

def normalize_image(image):
    return image / 255.0  # Normalize pixel values to [0, 1]

def flatten_image(image):
    return image.flatten()  # Convert (224, 224, 3) → (224 * 224 * 3,)

#def blur_image(image):
#    return cv2.GaussianBlur(image, (5, 5), 0)

# Define the pipeline
image_pipeline = Pipeline([
    ('resize', FunctionTransformer(resize_image, validate=False)),
    #('blur', FunctionTransformer(blur_image, validate=False)),
    ('normalize', FunctionTransformer(normalize_image, validate=False)),
    ('flatten', FunctionTransformer(flatten_image, validate=False))  # Flatten here
])
```

```python
# modify the createSampleDataset function to have an argument for data augmentation
# this includes flipping, rotating and blurring the images to get duplicates

def createSampleDataSet(sub_folder_paths, iterations = 4, augmented = False):
    training_data_numpy = []

    for i in range (iterations):
        sub_folder = sub_folder_paths[i]
        for dirpath, dirnames, filenames in os.walk(sub_folder): #walk through sub folder
            for file in filenames: #loop through files in sub folder  
                image_file_path = dirpath + "/" + file
                
                image = image_pipeline.fit_transform(image_file_path)
                training_data_numpy.append(image)
                
                if augmented == True:
                    augmented_images = augmentImage(image_file_path)
                    training_data_numpy.extend(augmented_images)
                    
    return training_data_numpy

def augmentImage(image_path):
    augmented_images = []
    
    base_image = Image.open(image_path).resize((224, 224))
    
    image_np = np.array(base_image)

    flipped_horiz = base_image.transpose(Image.FLIP_LEFT_RIGHT)  # Horizontal flip
    flipped_vert = base_image.transpose(Image.FLIP_TOP_BOTTOM)  # Vertical flip

    rotated_30 = base_image.rotate(90)  # Rotate 30 degrees
    rotated_neg180 = base_image.rotate(180)  
    rotated_neg270 = base_image.rotate(270)

    blurred_np = cv2.GaussianBlur(image_np, (5, 5), 0)  # Apply blur
    blurred = Image.fromarray(blurred_np)  # Convert back to PIL image

    augmented_images.extend(
        [flipped_horiz, flipped_vert, rotated_30, blurred, rotated_neg180, rotated_neg270]
    )
    
    #process images before returning
    augmented_images_np = [np.array(img, dtype=np.float32) / 255.0 for img in augmented_images]

    flattened_images = [img.flatten() for img in augmented_images_np]

    return np.array(flattened_images) 
    
def createLabelDataSet(sub_folder_paths, iterations = 4, augmented = False):
    data_numpy_y = []
    
    for i in range(iterations):
        label = os.path.basename(sub_folder_paths[i])
        num_instances_in_sub_folder = get_directory_size(sub_folder_paths[i])
        if augmented == True:
            # augmented adds 7 extra images
            num_instances_in_sub_folder = num_instances_in_sub_folder * 7
        data_numpy_y.extend([label] * num_instances_in_sub_folder)
            
    return data_numpy_y
```

```python
import os 
import sys
import numpy as np

def createSampleDataSet(sub_folder_paths, iterations, augmented = False):
    local_path_base = "/home/ec2-user/indian_birds_processed_images"
    
    for i in range (iterations):
        training_data_numpy = []
        last_two_folders = os.path.join(*sub_folder_paths[i].split(os.sep)[-2:])
        sub_folder = sub_folder_paths[i]

        for dirpath, dirnames, filenames in os.walk(sub_folder): #walk through sub folder
            for file in filenames: #loop through files in sub folder  
                image_file_path = dirpath + "/" + file
                
                image = image_pipeline.fit_transform(image_file_path)
                training_data_numpy.append(image)
                
                if augmented == True:
                    augmented_images = augmentImage(image_file_path)
                    training_data_numpy.extend(augmented_images)
        
        # store the images in the filesystem and release the memory
        
        full_local_path = local_path_base + "/" + last_two_folders
        if not os.path.exists(full_local_path):
            os.makedirs(full_local_path)
        
        file_path = os.path.join(full_local_path, "data.npy")
        np.save(file_path, training_data_numpy)
                    
        #release memory
        del training_data_numpy

    # read the first folder from the local file system, load it and return it
    
    #first_folder = os.path.join(*sub_folder_paths[0].split(os.sep)[-2:])
    #load_data_file_path = os.path.join(local_path_base, first_folder, "data.npy")
    #loaded_data = np.load(load_data_file_path)
    #return loaded_data

def augmentImage(image_path):
    augmented_images = []
    
    base_image = Image.open(image_path).resize((224, 224))
    
    image_np = np.array(base_image)

    flipped_horiz = base_image.transpose(Image.FLIP_LEFT_RIGHT)  # Horizontal flip
    flipped_vert = base_image.transpose(Image.FLIP_TOP_BOTTOM)  # Vertical flip

    rotated_30 = base_image.rotate(90)  # Rotate 30 degrees
    rotated_neg180 = base_image.rotate(180)  
    rotated_neg270 = base_image.rotate(270)

    blurred_np = cv2.GaussianBlur(image_np, (5, 5), 0)  # Apply blur
    blurred = Image.fromarray(blurred_np)  # Convert back to PIL image

    augmented_images.extend(
        [flipped_horiz, flipped_vert, rotated_30, blurred, rotated_neg180, rotated_neg270]
    )
    
    #process images before returning
    augmented_images_np = [np.array(img, dtype=np.float32) / 255.0 for img in augmented_images]

    flattened_images = [img.flatten() for img in augmented_images_np]

    return np.array(flattened_images) 
    
def createLabelDataSet(sub_folder_paths, iterations, augmented = False):
    local_path_base = "/home/ec2-user/indian_birds_processed_images"
        
    for i in range(iterations):
        last_two_folders = os.path.join(*sub_folder_paths[i].split(os.sep)[-2:])
        sub_folder = sub_folder_paths[i]
        
        data_numpy_y = []
        
        label = os.path.basename(sub_folder_paths[i])
        num_instances_in_sub_folder = get_directory_size(sub_folder_paths[i])
        if augmented == True:
            # augmented adds 7 extra images
            num_instances_in_sub_folder = num_instances_in_sub_folder * 7
        data_numpy_y.extend([label] * num_instances_in_sub_folder)
             
        #save numpy locally
        full_local_path = local_path_base + "/" + last_two_folders
        if not os.path.exists(full_local_path):
            os.makedirs(full_local_path)
        
        file_path = os.path.join(full_local_path, "labels.npy")
        np.save(file_path, data_numpy_y)
        
        del data_numpy_y

def get_directory_size(directory):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(directory):
        for filename in filenames:
            file_path = os.path.join(dirpath, filename)
            if os.path.isfile(file_path):  # Check if it is a file
                total_size += 1
    return total_size
```

```python
# train a model on each folder and save it
from sklearn.ensemble import RandomForestClassifier
import joblib 

# get directory size and loop through the directory getting all folders
directory_path = "/home/ec2-user/indian_birds_processed_images/train"
dir_size = get_directory_size(directory_path)
folders = [f for f in os.listdir(directory_path) if os.path.isdir(os.path.join(directory_path, f))]

for i in range(dir_size - 1):
    folder = folders[i]  # Get the folder at index `i`
    data_path = directory_path + "/" + folder + "/data.npy"
    label_path = directory_path + "/" + folder + "/labels.npy"
    loaded_data = np.load(data_path)
    loaded_labels = np.load(label_path)
    clf = RandomForestClassifier(n_estimators=100, random_state=42, oob_score=True, n_jobs=-1)
    clf.fit(loaded_data, loaded_labels)
    print(f"oob score is {clf.oob_score_}")
    
    # delete images from memory
    del loaded_data
    
    # save model
    model_dir = directory_path + "/" + folder
    model_path = os.path.join(model_dir, "random_forest.pkl")
    joblib.dump(clf, model_path)
    
    # delete model from memory
    del clf
```
